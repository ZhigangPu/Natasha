```python
def readline(filename):
    """Suppose every record occupies single line in the file, and this function sequeeze those records in a list
    
    Args:
        filename(str): filename of the data file, suppose data is stored in string format
    
    Returns:
        data(list[float]): list containing each record
    """
    data = []
    with open(filename, 'r') as f:
        for line in f:
            data.append(float(line.strip()))
    
    return data
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

TRAIN_LOSS_FILE = '../result/full/train_loss.txt'
VAL_LOSS_FILE = '../result/full/val_loss.txt'

# hyperparams
plt.rcParams['figure.figsize'] = [10, 5]
# data pts
train_loss = readline(TRAIN_LOSS_FILE)
val_loss = readline(VAL_LOSS_FILE)
x = [ i + 1 for i in list(range(len(train_loss)))]
# text
plt.xlabel('steps(2000iters)')
plt.ylabel('loss(cross entropy)')
plt.title('Loss graph')
plt.grid()
# execute
plt.plot(x, train_loss, color='blue', label='train')
plt.plot(x, val_loss, color='green', label='validation')
plt.legend()
```
